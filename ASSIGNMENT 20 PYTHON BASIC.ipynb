## 1. Set the variable test1 to the string 'This is a test of the emergency text system,' and save test1 to a file named test.txt.

Ans:-Here's how you can save the string 'This is a test of the emergency text system' to a file named 'test.txt':

test1 = 'This is a test of the emergency text system'

with open('test.txt', 'w') as file:

    file.write(test1)

In this code, we first define the string 'This is a test of the emergency text system' and assign it to the variable test1.

We then use the with statement to open a file named 'test.txt' in write mode ('w') and assign it to the variable file. This will automatically close the file when the block is exited.

We use the write() method of the file object to write the contents of test1 to the file.

After this code is executed, the string 'This is a test of the emergency text system' will be saved to a file named 'test.txt'.

## 3. Create a CSV file called books.csv by using these lines:

## 4. Use the sqlite3 module to create a SQLite database called books.db, and a table called books with these fields: title (text), author (text), and year (integer).

Ans:-To create a SQLite database called 'books.db' with a table called 'books' and the specified fields, you can use the following code:

import sqlite3

conn = sqlite3.connect('books.db')

c = conn.cursor()

create_table = '''CREATE TABLE books (

                    title TEXT,
                    
                    author TEXT,
                    
                    year INTEGER
                    
                );'''

c.execute(create_table)

conn.commit()

conn.close()


After this code is executed, a SQLite database called 'books.db' will be created with a table called 'books' and the specified fields.

## 5. Read books.csv and insert its data into the book table.

Ans:-To read the data from the CSV file 'books.csv' and insert it into the 'books' table in the SQLite database 'books.db', you can use the following code:

import sqlite3

import csv

conn = sqlite3.connect('books.db')

c = conn.cursor()

with open('books.csv', 'r') as f:

    reader = csv.reader(f)
    
    next(reader)  # skip the header row

    # Iterate over each row in the CSV file and insert it into the database
    for row in reader:
    
        title = row[0]
        
        author = row[1]
        
        year = int(row[2])
        
        c.execute("INSERT INTO books VALUES (?, ?, ?)", (title, author, year))

conn.commit()

conn.close()


After this code is executed, the data from the 'books.csv' file will be inserted into the 'books' table in the 'books.db' database.



## 6. Select and print the title column from the book table in alphabetical order.

## 8. Use the sqlalchemy module to connect to the sqlite3 database books.db that you just made in exercise 6.

Ans:-Here's an example of how to use the sqlalchemy module to connect to the sqlite3 database books.db:

from sqlalchemy import create_engine

engine = create_engine('sqlite:///books.db', echo=True)

result = engine.execute("SELECT * FROM books")

rows = result.fetchall()

for row in rows:

    print(row)

engine.dispose()

## 9. Install the Redis server and the Python redis library (pip install redis) on your computer. Create a Redis hash called test with the fields count (1) and name ('Fester Bestertester'). Print all the fields for test.

Ans:-To create a Redis hash called "test" with fields count (1) and name ('Fester Bestertester') using the Python Redis library, you can use the following code:

import redis

r = redis.Redis(host='localhost', port=6379, db=0)

r.hset('test', 'count', 1)

r.hset('test', 'name', 'Fester Bestertester')

print(r.hgetall('test'))


This will create a Redis hash called "test" with two fields, "count" and "name", with values 1 and "Fester Bestertester" respectively. The hgetall() method is then used to retrieve all fields and values for the hash, which are printed to the console.
